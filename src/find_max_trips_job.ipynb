# Find Max Trips Job

This notebook finds the day with the most unique trips from a view.

```python
# Get parameters from job widgets
dbutils.widgets.text("catalog", "")
dbutils.widgets.text("schema", "")
dbutils.widgets.text("view_name", "taxi_unique_trips")

catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
view_name = dbutils.widgets.get("view_name")
```

```python
# Set the default catalog and schema
spark.sql(f"USE CATALOG {catalog}")
spark.sql(f"USE SCHEMA {schema}")
```

```python
# Import the taxis module from the doan_dabs_test package
from doan_dabs_test import taxis

# Find the day with the most trips
result = taxis.find_day_with_most_trips(view_name)
```

```python
# Print the results
print(f"\n{'='*60}")
print(f"Day with Most Trips")
print(f"{'='*60}")
print(f"Date: {result['trip_date']}")
print(f"Unique Trip Count: {result['unique_trip_count']}")
print(f"{'='*60}\n")
```
